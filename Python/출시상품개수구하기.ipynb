```python
prog = [
    [93, 30, 55],
    [95, 90, 99, 99, 80, 99]
]

sp = [
    [1, 30, 5],
    [1,1,1,1,1,1]
]

def solution(progresses, speeds):
  #-------수행 완료 날짜 구하기-------
  pd=[] # Performance Date = '각 프로그램 작업완료까지 걸리는 시간'에 대한 리스트
  for i in range(len(progresses)):
    p, s = progresses[i], speeds[i]  # 값 초기화

    re = int((100 - p) / s) # 완료 날짜 구하기
    if (100 - p) % s != 0: #
      re += 1
    pd.append(re) # 임시저장


  #-------출시 상품 개수 구하기-------
  pivot = 0 # 비교 기준날짜
  answer = [1] # 한 날에 배포 가능한 각 물품의 수
  pd_i, answer_i = 1, 0 # 전체 반복문 인덱스 && answer 리스트의 인덱스
  
  while pd_i < len(pd):
    if pd[pivot] >= pd[pd_i]: # 기준날짜보다 짧은 작업일인 경우, '같은 출시일'
      answer[answer_i] += 1
    else:                     # 기준날짜보다 긴 작업일인 경우, '다른 출시일'
      pivot = pd_i # new기준날짜 갱신
      answer.append(1) # new기준날짜에 출시가능한 물품 수 갱신
      answer_i += 1 # answer리스트 수정으로 인한 인덱스 갱신

    pd_i += 1 # 전체 반복문 인덱스 증가
  return answer

an=solution(prog[0], sp[0])
print(an)
an=solution(prog[1], sp[1])
print(an)
```
